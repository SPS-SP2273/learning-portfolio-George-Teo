```python
from matplotlib import pyplot as plt
```

# Exercise 1

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',         # The 'o' marker refers to it being a circle
             ecolor='black',capsize=5,                     # ecolour refers to the colour of the error bars
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend('X vs Y')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
```

```python
y = []
for x in range(0, 51):
    y.append((np.exp(-x/10))*np.sin(x))
print(y)
```

```python
x = []
for n in range(0, 51):
    x.append(n)
print(x)
```

```python
plt.plot(x, y)
plt.xlabel('X', fontsize=15, color='darkorange')
plt.ylabel('Y', fontsize=15, color='darkorange', rotation=0)
plt.title(r'$y = e^{-x/10}sin(x)$', fontsize=20)
plt.grid(alpha=.25)
plt.ylim(-1,1)

# I am unsure how to make the graph smoother. I have attempted using x = np.linspace(0, 50, 50000) but I get
# ValueError: x and y must have same first dimension, but have shapes (50000,) and (50,)
```
